```python
def interval_search(file,inputinterval,direction='up',skip=0):
    
    '''
    DOCSTRING: Function takes in a file of a fugue score, an interval and a direction (up or down), and searches each individual voice for every melodic occurrence of that particular interval, at each starting pitch of the chromatic scale (distinguishing for enharmonically equivalent pitches). The output is a list of all such occurrences, organized by the starting pitch of the interval, with occurrences listed by voices (soprano, alto, tenor, bass). An empty bracket [] indicates that there were no occurrences found; if an occurence was noted, the number provided corresponds to the measure number containing the interval. 
    
    INPUT: File of a fugue (run code in cell above to get file names for fugues from WTC I and II), interval (interval name taken as a string, e.g. 'P5','m3','M3','d7','A4'), direction (default 'up' for ascending, can be set to 'down' for descending), skip (default set to 0, assuming that the interval is continguous and the notes are directly adjacent. To search for instances of outlines of noncontiguous intervals with one note interleaved, set value to 1 for this parameter).
    
    OUTPUT: Dictionary of the measure numbers of interval occurrences sorted by starting pitch; list within pitch is sorted by voice, from high (soprano) to low (bass).
    '''
    
    from collections import OrderedDict
    
    f=converter.parse(file)
    
    partList=[]
    
    for num in range(len(f.parts)):
        partList.append(f.parts[num].recurse().notes)
    
    startingpitch=['E', 'B', 'C-', 'F#', 'G-', 'C#', 'D-', 'G#', 'A-', 'D#', 'E-', 'A#', 'B-', 'E#', 'F', 'B#', 'C', 'G', 'D', 'A'] #work on enharmonic pitches
    
    i=interval.Interval(str(inputinterval))
     
    if direction=='down': 
        i=i.reverse()
    
    if abs(i.semitones)>12:
        compound=True
    else:
        compound=False
    
    searchLists=[]
    for mypitch in startingpitch:
        endnote=note.Note(mypitch).transpose(i)
        if skip==0:
            searchLists.append([note.Note(mypitch),endnote])
        elif skip==1:
            searchLists.append([note.Note(mypitch),search.Wildcard(),endnote])

    totallist=[]

    for searchList in searchLists:

        allpList=[]
        for pList in partList:
            mlst=search.noteNameSearch(pList,searchList)
            mlist=[]
            for i in mlst:
                if skip==0:
                    if direction=='up':
                        if compound==False:
                            if pList[i].pitch.frequency<pList[i+1].pitch.frequency:
                                mlist.append(pList[i].measureNumber)
                        elif compound==True:
                            if pList[i].pitch.frequency<pList[i+1].pitch.frequency and abs(interval.Interval(pList[i],pList[i+1]).semitones)>12:
                                mlist.append(pList[i].measureNumber)
                    elif direction=='down':
                        if compound==False:
                            if pList[i].pitch.frequency>pList[i+1].pitch.frequency:
                                mlist.append(pList[i].measureNumber)
                        elif compound==True:
                            if pList[i].pitch.frequency>pList[i+1].pitch.frequency and abs(interval.Interval(pList[i],pList[i+1]).semitones)>12:
                                mlist.append(pList[i].measureNumber)
                elif skip==1:
                    if direction=='up':
                        if compound==False:
                            if pList[i].pitch.frequency<pList[i+2].pitch.frequency:
                                mlist.append(pList[i].measureNumber)
                        elif compound==True:
                            if pList[i].pitch.frequency<pList[i+2].pitch.frequency and abs(interval.Interval(pList[i],pList[i+2]).semitones)>12:
                                mlist.append(pList[i].measureNumber)
                    elif direction=='down':
                        if compound==False:
                            if pList[i].pitch.frequency>pList[i+2].pitch.frequency:
                                mlist.append(pList[i].measureNumber)
                        elif compound==True:
                            if pList[i].pitch.frequency>pList[i+2].pitch.frequency and abs(interval.Interval(pList[i],pList[i+2]).semitones)>12:
                                mlist.append(pList[i].measureNumber)
            allpList.append(mlist)
        totallist.append(allpList)
    
    totaldict=OrderedDict(zip(startingpitch,totallist))

    return totaldict
```
